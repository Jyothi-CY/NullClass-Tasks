```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import re
from collections import Counter
from wordcloud import WordCloud

# Load your filtered datasets
health_fitness_reviews = pd.read_csv('health_fitness_reviews.csv')
five_star_reviews = pd.read_csv('health_fitness_5star_reviews.csv')
```

```python
def prepare_wordcloud_data(reviews_df):
    app_names = set(reviews_df['App'].str.lower().unique())
    
    lemmatizer = WordNetLemmatizer()
    custom_stopwords = set(stopwords.words('english')).union({
        'app', 'apps', 'application', 'use', 'using', 'thank', 'thanks',
        'great', 'good', 'nice', 'like', 'really', 'would', 'could'
    })
    
    text = ' '.join(reviews_df['Translated_Review'].astype(str).str.lower())
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    words = word_tokenize(text)
    processed_words = [lemmatizer.lemmatize(word) for word in words
        if word not in custom_stopwords
        and word not in app_names
        and len(word) > 2
        and not word.isnumeric()]
    
    word_counts = Counter(processed_words)
    word_counts['yoga'] = word_counts.get('yoga', 0) * 1.5
    word_counts['workout'] = word_counts.get('workout', 0) * 1.3
    
    return word_counts

# Generate word frequencies
word_freq = prepare_wordcloud_data(five_star_reviews)    
word_freq
```

```python
# Create professional word cloud with custom colors
plt.figure(figsize=(16, 12))
wordcloud = WordCloud(
    width=1400,
    height=1000,
    background_color='white',
    colormap='plasma',
    max_words=200,
    contour_width=2,
    contour_color='steelblue',
    collocations=False,
    prefer_horizontal=0.8).generate_from_frequencies(word_freq)
```

```python
plt.imshow(wordcloud, interpolation='gaussian')  
plt.axis('off')
plt.title('Top Keywords in 5-Star Health & Fitness Reviews\n(With Term Boosting)', 
          fontsize=20, pad=25, fontweight='bold')
plt.tight_layout()

plt.savefig('health&fitness_wordcloud_enhanced.png', dpi=350, bbox_inches='tight')
```

```python
# Print top keywords
print("Top 20 Keywords:")
for word, freq in word_freq.most_common(20):
    print(f"{word}: {freq}")
```

```python
# Display top keywords with formatting

print("TOP 20 KEYWORDS (WITH FREQUENCY BOOSTING) : ")
print("\n")
for i, (word, freq) in enumerate(word_freq.most_common(20), 1):
    print(f"{i:>2}. {word.upper():<15} {freq:>5} mentions")
print("\nWord cloud saved as 'health_fitness_wordcloud_enhanced.png'")
plt.show()
```
