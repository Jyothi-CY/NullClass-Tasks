```python
# Data Cleaning

apps_df = apps_df.dropna(subset=['Rating'])
for column in apps_df.columns:
    apps_df[column].fillna(apps_df[column].mode()[0])
apps_df.drop_duplicates   
apps_df = apps_df[apps_df['Rating']<=5]
reviews_df.dropna(subset=['Translated_Review'],inplace=True)
```

```python
# First check the current data type
print(f"Installs dtype before conversion: {apps_df['Installs'].dtype}")
print(f"Price dtype before conversion: {apps_df['Price'].dtype}")

# Convert only if the column contains strings
if apps_df['Installs'].dtype == 'object':
    apps_df['Installs'] = apps_df['Installs'].str.replace('[+,]', '', regex=True).astype(int)
else:
    print("Installs column is already numeric")

if apps_df['Price'].dtype == 'object':
    apps_df['Price'] = apps_df['Price'].str.replace('$', '', regex=False).astype(float)
else:
    print("Price column is already numeric")
```

```python
def clean_installs(val):
    if isinstance(val, str):
        return int(val.replace(',', '').replace('+', ''))
    return val

def clean_price(val):
    if isinstance(val, str):
        return float(val.replace('$', ''))
    return val

apps_df['Installs'] = apps_df['Installs'].apply(clean_installs)
apps_df['Price'] = apps_df['Price'].apply(clean_price)

# Verify conversion
print("\nAfter conversion:")
print(f"Installs dtype: {apps_df['Installs'].dtype}")
print(f"Price dtype: {apps_df['Price'].dtype}")
```

```python
# For missing Android versions, we can use a placeholder like "Not specified"
apps_df['Android Ver'] = apps_df['Android Ver'].fillna('Not specified')

# For missing Current Ver, we can either drop these rows or fill with a placeholder
apps_df['Current Ver'] = apps_df['Current Ver'].fillna('Not specified')
```

```python
import tkinter as tk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from datetime import datetime, time
```

```python
# Apply filters
filtered_df = apps_df[
    (apps_df['Installs'] >= 10000) &
    (apps_df['Revenue'] >= 10000) &
    (apps_df['Android Ver'] > 4.0) &
    (apps_df['Size'] > 15) &
    (apps_df['Content Rating'] == 'Everyone') &
    (apps_df['App Length'] <= 30)
]
```

```python
# Determine top 3 categories by number of apps after filtering
top_categories = filtered_df['Category'].value_counts().nlargest(3).index.tolist()
filtered_df = filtered_df[filtered_df['Category'].isin(top_categories)]
```

```python
# Create root window
root = tk.Tk()
root.title("Time-Sensitive Chart")

# Time check: only allow chart between 1 PM and 2 PM IST
india_time = datetime.now(pytz.timezone("Asia/Kolkata")).time()
allowed_start = time(13, 0)
allowed_end = time(14, 0)
```

```python
if allowed_start <= india_time <= allowed_end:
    # Group by Category and Type
    group_data = filtered_df.groupby(['Category', 'Type']).agg({
        'Installs': 'mean',
        'Revenue': 'mean'
    }).reset_index()

    # Get top 3 categories by total installs
    top3_categories = group_data.groupby('Category')['Installs'].sum().nlargest(3).index
    group_data = group_data[group_data['Category'].isin(top3_categories)]

    categories = sorted(group_data['Category'].unique())
    bar_width = 0.35
    index = list(range(len(categories)))

    fig, ax1 = plt.subplots(figsize=(12, 7))
    types = ['Free', 'Paid']
    colors = {'Free': 'skyblue', 'Paid': 'orange'}

    for i, t in enumerate(types):
        data = group_data[group_data['Type'] == t]
        data = data.set_index('Category').reindex(categories).reset_index()
        
        ax1.bar(
            [x + i * bar_width for x in index],
            data['Installs'].fillna(0),
            width=bar_width,
            label=f'Installs - {t}',
            color=colors[t],
            align='center'
        )

    ax2 = ax1.twinx()

    for i, t in enumerate(types):
        data = group_data[group_data['Type'] == t]
        data = data.set_index('Category').reindex(categories).reset_index()
        
        ax2.plot(
            [x + bar_width/2 for x in index],
            data['Revenue'].fillna(0),
            label=f'Revenue - {t}',
            marker='o',
            linestyle='--',
            color=colors[t]
        )

    ax1.set_xlabel('App Categories')
    ax1.set_ylabel('Average Installs')
    ax2.set_ylabel('Average Revenue ($)')
    ax1.set_title('Average Installs vs Revenue (Free vs Paid) in Top 3 Categories')
    ax1.set_xticks([x + bar_width/2 for x in index])
    ax1.set_xticklabels(categories)
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
    ax1.grid(True, linestyle='--', alpha=0.5)

    # Embed in Tkinter
    canvas = FigureCanvasTkAgg(fig, master=root)
    canvas.draw()
    canvas.get_tk_widget().pack()

else:
    label = tk.Label(root, text="Chart not displayed. Allowed time is between 1 PM and 2 PM IST.")
    label.pack()

root.mainloop()
```
